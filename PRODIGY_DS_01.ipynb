 Install & Import Libraries

```python
# These are pre-installed in Colab, just import them
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

 Create Sample Dataset (Age & Gender)

```python
# We'll create a realistic sample dataset
# (You can also download from the GitHub link in your task)

np.random.seed(42)
n = 1000  # 1000 people

data = {
    'Age': np.concatenate([
        np.random.normal(25, 5, 300),   # young adults
        np.random.normal(40, 8, 400),   # middle aged
        np.random.normal(60, 7, 300)    # seniors
    ]).astype(int),
    'Gender': np.random.choice(['Male', 'Female', 'Other'], size=1000, p=[0.49, 0.49, 0.02])
}

df = pd.DataFrame(data)

# Clip ages to realistic range
df['Age'] = df['Age'].clip(1, 90)

print("Dataset shape:", df.shape)
print(df.head())
print("\nGender counts:\n", df['Gender'].value_counts())
```

Cell 3 — Histogram of Age Distribution

```python
plt.figure(figsize=(10, 6))

plt.hist(df['Age'], bins=20, color='steelblue', edgecolor='black', alpha=0.8)

plt.title('Age Distribution of Population', fontsize=16, fontweight='bold')
plt.xlabel('Age', fontsize=13)
plt.ylabel('Number of People', fontsize=13)
plt.grid(axis='y', alpha=0.4)

plt.tight_layout()
plt.savefig('age_distribution.png', dpi=150)  # saves the chart
plt.show()
print("✅ Age histogram saved!")
```

 Bar Chart of Gender Distribution

```python
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(8, 5))

colors = ['#4C72B0', '#DD8452', '#55A868']
bars = plt.bar(gender_counts.index, gender_counts.values, color=colors, edgecolor='black', width=0.5)

# Add count labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 5,
             f'{int(height)}', ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.title('Gender Distribution of Population', fontsize=16, fontweight='bold')
plt.xlabel('Gender', fontsize=13)
plt.ylabel('Number of People', fontsize=13)
plt.grid(axis='y', alpha=0.4)

plt.tight_layout()
plt.savefig('gender_distribution.png', dpi=150)
plt.show()
print("✅ Gender bar chart saved!")
```

Age Distribution by Gender (Combined Chart)

```python
plt.figure(figsize=(12, 6))

colors = {'Male': '#4C72B0', 'Female': '#DD8452', 'Other': '#55A868'}

for gender, color in colors.items():
    subset = df[df['Gender'] == gender]['Age']
    plt.hist(subset, bins=20, alpha=0.6, label=gender, color=color, edgecolor='black')

plt.title('Age Distribution by Gender', fontsize=16, fontweight='bold')
plt.xlabel('Age', fontsize=13)
plt.ylabel('Number of People', fontsize=13)
plt.legend(fontsize=12)
plt.grid(axis='y', alpha=0.4)

plt.tight_layout()
plt.savefig('age_by_gender.png', dpi=150)
plt.show()
print("✅ Combined chart saved!")
```
